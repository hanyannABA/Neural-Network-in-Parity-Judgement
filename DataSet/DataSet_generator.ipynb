```python
import random
import pandas as pd

#生成一个数据集，参数包括数据集大小、最小值和最大值。每个数据点由一个随机整数x和一个标签y组成，其中y表示x的奇偶性（0表示偶数，1表示奇数）
def generate_dataset(size: int, min_value: int, max_value: int):
    data=[]
    for _ in range(size):
        x = random.randint(min_value, max_value)
        y = x%2
        data.append([x, y])

    return pd.DataFrame(data, columns=["x", "y"])



if __name__ == "__main__":
    #生成数据集
    train = generate_dataset(10000, 0, 10**10)
    validation = generate_dataset(1000, 0, 10**12)
    test = generate_dataset(1000, 0, 10**15)
    #将数据集保存为CSV文件
    train.to_csv("train.csv", index=False)
    validation.to_csv("validation.csv", index=False)
    test.to_csv("test.csv", index=False)
```
